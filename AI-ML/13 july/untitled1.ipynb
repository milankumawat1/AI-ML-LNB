```python
ds=pd.DataFrame()
ds
```

```python
for i in range(1,500):
    response=requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=71f66b2394d018b59baec60a7fbe8ba3&&page={}'.format(i))
    temp_df=pd.DataFrame(response.json()['results'])[['id','original_title','overview','popularity','release_date','title','vote_average','vote_count']]
    ds=ds._append(temp_df,ignore_index=True)
```

```python
ds.to_csv('MovieDataset.csv')
```

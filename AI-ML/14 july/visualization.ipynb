```python
index=[2008,2009,2010,2011,2012]
rohit=[404,362,464,372,432]
virat=[501,540,200,201,439]
```

```python
plt.plot(index,virat,color='red',label='Virat Kolhi')
plt.title('Virat Run')
plt.xlabel('Year')
plt.ylabel('Run')
plt.legend()
plt.show()

```

```python
plt.plot(index,rohit)
plt.title('Rohit Run')
plt.xlabel('Year')
plt.ylabel('Run')
```

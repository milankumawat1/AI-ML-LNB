Pie Chart:
    univariate/bivariate analysis
    categorical vs numerical
    use case: to find the contribution on a standard scale

```python
data=[25,45,100,20,40]
sub=['eng','sci','math','hin','bio']
plt.pie(data,labels=sub,autopct='%1.0f')
plt.show()
```

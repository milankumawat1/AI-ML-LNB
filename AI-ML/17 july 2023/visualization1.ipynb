Bar Chart :
    Bivariate Analysis
    numerical vs categorical
    Used case: Aggregate Analysis of the group

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
children=[10,20,30,40,50]
color=['red','blue','green','yellow','pink']
plt.bar(color,children,color='black')
plt.show()
```

```python
children=[10,20,30,40,50]
color=['red','blue','green','yellow','pink']
plt.barh(color,children,color='black')
plt.show()
```

```python
df=pd.read_csv('../Datasets/batsman_season_record.csv')
df
```

```python
plt.bar(df['batsman'],df['2015'])
plt.bar(df['batsman'],df['2016'])
plt.bar(df['batsman'],df['2017'])
plt.show()
```

```python
plt.bar(np.arange(df.shape[0])-0.2,df['2015'],width=0.2)
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.2)
plt.bar(np.arange(df.shape[0])+0.2,df['2017'],width=0.2)
```

```python
plt.bar(np.arange(df.shape[0])-0.2,df['2015'],width=0.2)
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.2)
plt.bar(np.arange(df.shape[0])+0.2,df['2017'],width=0.2)
plt.xticks(np.arange(df.shape[0]),(df['batsman']))
plt.show
```

```python
plt.bar(np.arange(df.shape[0])-0.2,df['2015'],width=0.2)
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.2)
plt.bar(np.arange(df.shape[0])+0.2,df['2017'],width=0.2)
plt.xticks(np.arange(df.shape[0]),(df['batsman']),rotation='vertical')
plt.show
```

```python
plt.barh(np.arange(df.shape[0])-0.2,df['2015'],height=0.2)
plt.barh(np.arange(df.shape[0]),df['2016'],height=0.2)
plt.barh(np.arange(df.shape[0])+0.2,df['2017'],height=0.2)
plt.yticks(np.arange(df.shape[0]),(df['batsman']))
plt.show
```

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
response=requests.get('https://www.ambitionbox.com/jobs/search?tag=data%20science',headers=headers).text
soup=BeautifulSoup(response,'html.parser')
soup
```

```python
company=soup.find_all('div',class_='jobInfoCard')
company
```

```python

company_name =[]
reviews=[]
no_of_years_exp=[]
packages=[]
location=[]
skill_requirement=[]
for i in company:
    company_name.append(i.find('p',class_='company').text.strip())
    reviews.append(i.find('a',class_='review-count').text.strip())
    entity=i.find_all('div',class_='entity')
    if len(entity)>0:
        no_of_years_exp.append(entity[0].text.strip())
    else:
        no_of_years_exp.append("N/A")
    if len(entity)>1:
        packages.append(entity[1].text.strip())
    else:
        packages.append("N/A")
    if len(entity)>2:
        location.append(entity[2].text.strip())
    else:
        location.append("N/A")
    if len(entity)>3:
        skill_requirement.append(entity[3].text.strip())
    else:
        skill_requirement.append("N/A")
    
    
    
df=pd.DataFrame({
    'Company Name':company_name,
    'Reviews':reviews,
    'No of years or Experience':no_of_years_exp,
    'Package':packages,
    'Location':location,
    'Skill Required':skill_requirement
})

df
```
